# 3.5 Gradient Descent 
![image.png](https://raw.githubusercontent.com/Pokemongle/img_bed_0/main/img/20251114213852.png)  
- step choice
    - $10^{\gamma}$, usually $\gamma$ is negative integer
    - diminishing steplength like $\alpha = \frac{1}{k}$
    > it's ok for cost to oscillate because it could hop around  
- convergence criteria  
    - $\|\nabla g(w^{k-1})\|_2$ is very small  
    - $\frac{1}{N}\|w^k - w^{k-1}\|_2 < \epsilon$
    - $\frac{1}{N}\|g(w^k) - g(w^{k-1})\|_2 < \epsilon$
    - **set max iteration manually**
        - smaller $\alpha$, bigger iter_max

```python
from autograd import grad

def gradient_descent(g, alpha, max_its, w):
    gradient = grad(g)

    weight_history = [w]
    cost_history = [g(w)]

    for k in range(max_its):
        grad_eval = gradient(w)
        w = w - alpha*grad_eval 
        weight_history.append(w)
        cost_history.append(g(w))
    
    return weight_history, cost_history 
```
