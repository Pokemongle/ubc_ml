# 1. head files

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn import svm
```

# 2. Preprocess dataset

```python
iris = datasets.load_iris()
```

```python
print(iris.feature_names)
print(iris.target_names)
```

```python
# get labels and features, use petal this time
mask = (iris.target == 1) | (iris.target == 2) # versicolor vs virginica
x1 = iris.data[mask, 2] # petal length
x2 = iris.data[mask, 3] # petal width
y = iris.target[mask]
x = np.array([x1, x2])
x = x.T
```

```python
plt.scatter(x[y==1,0],x[y==1,1],color='r')
plt.scatter(x[y==2,0],x[y==2,1],color='b')
plt.legend(['versicolor','virginica'])
plt.grid()
plt.xlim(2.5, 7.1)
plt.ylim(0.5, 3.0)
plt.xlabel('Petal length(cm)')
plt.ylabel('Petal width(cm)')
plt.title('Versicolor vs Virginica')
```

# 3. train model(Virginica vs Versicolor)

```python
# train model1
svm_model = svm.SVC(kernel='linear')
svm_model.fit(x,y)
```

```python
w1 = svm_model.coef_[0][0]
w2 = svm_model.coef_[0][1]
b = svm_model.intercept_
x1p = np.linspace(2.5,7.1,1000)
x2p = -(w1*x1p+b)/w2
```

```python
plt.scatter(x[y==1,0],x[y==1,1],color='r', label="Versicolor")
plt.scatter(x[y==2,0],x[y==2,1],color='b', label='Virginica')
plt.legend(['versicolor','virginica'])
plt.grid()
plt.xlim(2.5, 7.1)
plt.ylim(0.5, 3.0)
plt.plot(x1p, x2p, label=fr'$x_{2}={(-w1/w2).item():.4f}x_{1}+{(-b/w2).item():.4f}$')
plt.legend()
plt.xlabel('Petal length(cm)')
plt.ylabel('Petal width(cm)')
plt.title('Versicolor vs Virginica')
```

```python
# accuracy 
y_pred = svm_model.predict(x)
accuracy = (y_pred == y).mean()
print(f"Accuracy for Versicolor vs Virginica: {accuracy}")
```

# 4. Setosa vs Versicolor

## import dataset and visualization

```python
# get labels and features, use sepal this time
mask = (iris.target == 0) | (iris.target == 1)
x1 = iris.data[mask, 2]
x2 = iris.data[mask, 3]
y = iris.target[mask]
x = np.array([x1, x2])
x = x.T
# data visualization
plt.scatter(x[y==0,0],x[y==0,1],color='g')
plt.scatter(x[y==1,0],x[y==1,1],color='b')
plt.legend(['setosa','versicolor'])
plt.grid()
# plt.xlim(2.5, 7.1)
# plt.ylim(0.5, 3.0)
plt.xlabel('Petal length(cm)')
plt.ylabel('Petal width(cm)')
plt.title('Setosa vs Versicolor')
```

## train the svm model and calculate accuracy

```python
# train model1
svm_model = svm.SVC(kernel='linear')
svm_model.fit(x,y)
# accuracy 
y_pred = svm_model.predict(x)
accuracy = (y_pred == y).mean()
print(f"Accuracy for Versicolor vs Virginica: {accuracy}")
# visualization
w1 = svm_model.coef_[0][0]
w2 = svm_model.coef_[0][1]
b = svm_model.intercept_
x1p = np.linspace(0.5,5.5,1000)
x2p = -(w1*x1p+b)/w2
plt.scatter(x[y==0,0],x[y==0,1],color='g', label='Setosa')
plt.scatter(x[y==1,0],x[y==1,1],color='b', label='Versicolor')
plt.legend(['setosa', 'versicolor'])
plt.grid()
plt.xlim(0.5, 5.5)
plt.ylim(0.0, 2.0)
plt.plot(x1p, x2p, label=f'$x_2p={(-w1/w2).item():.4f}x_1+{(-b/w2).item():.4f}$')
plt.legend()
plt.xlabel('Petal length(cm)')
plt.ylabel('Petal width(cm)')
plt.title('Setosa vs Versicolor')
```
