# Linear Regression  
- Model: $y_p = w_0 + w_1x_{1,p} + w_2x_{2,p}+...+w_Nx_{N,p} \forall p$  
- Cost: Mean Square Error(MSE)
    $g(\bold{w})=\frac{1}{P}\Sigma_{p=1}^{P}(\bar{\bold{x}}_p^T\bold{w}-y_p)^2$ 
    - or Mean Absolute Deviation(MAD) if Outliers
    $g(\bold{w})=\frac{1}{P}\Sigma_{p=1}^{P}|\bar{\bold{x}}_p^T\bold{w}-y_p|$
- Evaluation
    - $R^2$ coefficient
        - $R^2 = 1-\frac{MSE}{V}$
        - V=$\frac{1}{P}\Sigma_{p=1}^{P}(y_p-\bar{y})^2$

- Principle: Gradient Descent / Newton's Method / sth else (see lg_byhand)
- Code: 
    - dataset: iris 
    - function: sklearn model.fit model.coef_ model.intercept_ model.predict model.score(see lg_scikit)
- Mathematical things 
    - Linear Regression in Closed-Form


```python
# Linear Regression in Scikit Learn
# ENGR 518 Lecture 3

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import datasets

iris = datasets.load_iris()

"""
print(iris.DESCR)
print(iris.data)
print(iris.feature_names)
print(iris.target)
print(iris.target_names)
"""
print(iris.feature_names)
print(iris.target_names)
x = iris.data[:,2].reshape(-1,1) # column selection, reshape to vector
y = iris.data[:,3].reshape(-1,1)

plt.scatter(x,y)
plt.grid(1)

# learn the linear regression model
# create and fit model
model = LinearRegression() # create model
model.fit(x,y) # fit model
print('intercept: ', model.intercept_, 'coef: ', model.coef_) # learned parameters
print(model.score(x, y))
plt.plot(x, model.predict(x), 'r') # how to draw the learned line
plt.xlabel('petal length')
plt.ylabel('petal width')
```

- why use d as x is better?
    - because we only assume y has error, and in the real world, the error of detecting t is much bigger than d

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

d = np.array([
    100,100,100,127,127,127,152,152,152,178,178,178
]).reshape(-1,1)
t2 = np.array([
    0.36,0.38,0.46,0.46,0.49,0.51,0.50,0.53,0.56,0.55,0.58,0.61
]).reshape(-1,1)
model = LinearRegression()
model.fit(d, t2)
coef = model.coef_[0][0]
intercept = model.intercept_[0]
print(coef, intercept)
print(f'g={2/coef/100} m/s^2')
```
