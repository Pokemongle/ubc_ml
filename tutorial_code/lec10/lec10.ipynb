```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# generate random linearly distributed data
P = 10
x = np.linspace(0, 10, P)
# LinearRegression expects N-D array
x = np.reshape(x, [P, 1])  # -1 to autocalculate n. cols
noise = 5 * np.random.randn(P, 1)
y = 1 + x + x**2 + noise
plt.scatter(x, y)
plt.grid(1)

# generate polynomial features
d = 2
# generate polynomial function of degree =
polynomial_features = PolynomialFeatures(d)

# generate features using function
x_poly = polynomial_features.fit_transform(x)

# learn the linear regression model on the poly features
# create and fit model
model = LinearRegression()
model.fit(x_poly, y)

plt.plot(x, model.predict(x_poly), 'r')
plt.xlabel('x')
plt.ylabel('y')

print(model.intercept_, model.coef_)
```
